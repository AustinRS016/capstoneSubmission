```python
import pandas as pd
esc = pd.read_csv('wdfwHatchery6_18.csv', low_memory=False)
esc = esc[['Species', 'Event','Origin','Brood Year','Run','Facility','Adult Count','Date']]
esc = esc[esc.Event == 'Trap Estimate']
esc = esc.drop(columns=['Event'])
esc = esc[esc['Adult Count'] != 0]
esc['Date'] = pd.to_datetime(esc['Date'])
esc['DOY'] = esc['Date'].dt.strftime('%j')
esc['Year'] = esc['Date'].dt.strftime('%Y')
esc['Year'] = esc['Year'].astype(str).astype(int)
esc
```

```python
species = ['Chinook','Coho','Sockeye','Pink','Chum','Steelhead']
esc = esc[esc.Species.isin(species)]
esc.info()
```

```python
#Create lists for Facilities, Runs, Brood Years, Origin, Species
run = esc[['Run']].drop_duplicates()
run = run.values.tolist()
# brood = esc[['Brood Year']].drop_duplicates()
# brood = brood.values.tolist()
origin = [['WILD'],['HATCHERY'],['UNKNOWN']]
facilities = esc[['Facility']].drop_duplicates()
facilities = facilities.values.tolist()
species = esc[['Species']].drop_duplicates()
species = species.values.tolist()
facilities = sorted(facilities)
```

```python
i = 0
l = []
for f in facilities:
    time = esc[esc.Facility.isin(f)]
    if time.Year.isin([2020]).any().any():
        l.append(f)
        

        i+= 1


print(i)
facilities = l
facilities
```

```python
k = 0
#for each facility in study
tst = []
for f in facilities:
    fac = esc[esc.Facility.isin(f)]
    if fac.Year.isin([2020]).any().any():
    
        #for each species in each facility
        for s in species:
            facspec = fac[fac.Species.isin(s)]
            if facspec.empty == True:
                continue

            #for if the species for each facility is wild, hatchery or unknown
            for org in origin:
                facspecorg = facspec[facspec.Origin.isin(org)]
                if facspecorg.empty == True:
                    continue

                #for each run for each species in the facility
                for r in run:
                    final = facspecorg[facspecorg.Run.isin(r)]
                    if final.empty == True:
                        continue
                    if final.shape[0] > 30:
                        k += 1
                        
                        if f not in tst:
                            tst.append(f) 
    else:
        continue
# tst.to_csv('facilities_inStudy.csv', index=False)
print(len(tst))
facilities = tst
tst
```

```python
for f in facilities:
    finaldf = pd.DataFrame()
    fac = esc[esc.Facility.isin(f)]
    for s in species:
        facspec = fac[fac.Species.isin(s)]
        if facspec.empty == True:
            continue
        for org in origin:
            facspecorg = facspec[facspec.Origin.isin(org)]
            if facspecorg.empty == True:
                continue
            for r in run:
                final = facspecorg[facspecorg.Run.isin(r)]
                if final.empty == True:
                    continue
                if final.shape[0] > 30:
                    i += 1

                    #query hatchery, species; reduce columns
                    df = final[['Adult Count','DOY','Year']]

                    #Create row for each day and reduce to minimize file size
                    yearSum = df.groupby('Year').sum('Adult Count')
                    df = df.merge(yearSum, on='Year')

                    df = df.rename(columns = {'Adult Count_x':'Count', 'Adult Count_y':'yearTotal'})
                    df['%total'] = df['Count']/df.yearTotal
                    df = df.groupby('DOY').sum('%total')

                    df['dens'] = ((df['%total']*100))
                    df = df.round({'dens':0})
                    df = df.reset_index()
                    print(f"Computed density for {f[0]} {org[0]} {r[0]} {s[0]}")
                    
                    colname=(org[0].capitalize()+" "+str(r[0]).replace("nan","")+" "+s[0]).replace("  "," ")
                    
                    xf = pd.DataFrame(columns=[colname])

                    for index, row in df.iterrows():
                        i = 1
                        while i <= row['dens']:
                            xf = xf.append({colname:row['DOY']},ignore_index=True)
                            i += 1
                    finaldf = pd.concat([finaldf, xf], axis=1)



    
    print(finaldf)
    finaldf.to_csv(f'densityPlotData/{f[0].replace(" ","_")}.csv',index=False)

```
